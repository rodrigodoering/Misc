#### Importa todas as bibliotecas e o script

```python
import pandas as pd
import numpy as np
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from script_objeto import WebMining
```

#### Executa o processo de WebScrapping e retorna o dataframe normalizado

```python
crawler = WebMining('to_craw.txt') 
df = crawler.ExecutarCrawler(',', True)
```

#### Gera o dataframe NÃO normalizado para visualização

```python
df_original = crawler.ExecutarCrawler(',')
df_original
```

```python
df
```

```python
MaxHomicidio, MinHomicidio = df_original['TaxaHomicidio'].max(), df_original['TaxaHomicidio'].min()
MaxPobreza, MinPobreza = df_original['IndicePobreza'].max(), df_original['IndicePobreza'].min()
MaxSeguranca, MinSeguranca = df_original['InvestimentoSeguranca'].max(), df_original['InvestimentoSeguranca'].min()
MaxGINI, MinGINI = df_original['GINI'].max(), df_original['GINI'].min()

coefHomicidio = [MaxHomicidio, MinHomicidio]
coefPobreza = [MaxPobreza, MinPobreza]
coefSeguranca = [MaxSeguranca, MinSeguranca]
coefGINI = [MaxGINI, MinGINI]

matrizCoef1 = [coefHomicidio, coefPobreza, coefSeguranca]
matrizCoef2 = [coefHomicidio, coefGINI, coefSeguranca]
```

```python
def reverter(valor, Max, Min):
    return (valor*Max) - (valor*Min) + Min
```

#### Clusterização usando o Taxa de homicidio, investimento em segurança e índice de pobreza

```python

dataset = df[['TaxaHomicidio','IndicePobreza','InvestimentoSeguranca']]
cluster = KMeans(n_clusters=2, random_state=42).fit(dataset)
classes_k2 = cluster.labels_
score = silhouette_score (dataset, classes_k2, metric='euclidean')
print(score)
```

```python

dataset = df[['TaxaHomicidio','IndicePobreza','InvestimentoSeguranca']]
cluster = KMeans(n_clusters=3, random_state=42).fit(dataset)
classes_k3 = cluster.labels_
score = silhouette_score (dataset, classes_k3, metric='euclidean')
print(score)
```

```python

dataset = df[['TaxaHomicidio','IndicePobreza','InvestimentoSeguranca']]
cluster = KMeans(n_clusters=4, random_state=42).fit(dataset)
classes_k4 = cluster.labels_
score = silhouette_score (dataset, classes_k4, metric='euclidean')
print(score)
```

```python
dataset['cluster'] = classes_k2
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.array(dataset['TaxaHomicidio'])
y = np.array(dataset['IndicePobreza'])
z = np.array(dataset['InvestimentoSeguranca'])
ax.set_xlabel('TaxaHomicidio')
ax.set_ylabel('IndicePobreza')
ax.set_zlabel('InvestimentoSeguranca')
ax.scatter(x,y,z, c=dataset["cluster"], s=70, alpha=1)
plt.gcf().subplots_adjust(bottom=0.01)
plt.show()
print(dataset['cluster'])
```

```python
centro = cluster.cluster_centers_
for i in range(len(centro)):
    for j in range(len(centro[i])):
        centro[i][j] = reverter(centro[i][j], matrizCoef1[j][0], matrizCoef1[j][1])
centro
```

#### Clusterizações específicas

```python
df1 = df[['TaxaHomicidio', 'GINI']]

cluster = KMeans(n_clusters=3, random_state=42).fit(df1)
classes = cluster.labels_
centers = cluster.cluster_centers_
df_classes = pd.DataFrame(data=cluster.labels_, index=crawler.estados, columns=['Cluster'])
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(df['TaxaHomicidio'],df['GINI'],c=classes,s=50)
for i,j in centers:
    ax.scatter(i,j,s=50,c='red',marker='+')
ax.set_xlabel('TaxaHomicidio')
ax.set_ylabel('GINI')
plt.colorbar(scatter)

fig.show()
df_classes
```
